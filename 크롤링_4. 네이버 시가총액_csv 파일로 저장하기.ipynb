## 네이버 시가총액 정보를 가져와서 csv 파일로 저장하기

```python
import csv
import requests
from bs4 import BeautifulSoup
```

### for 문으로 처리하기
1. table 안에 있는 tbody 만 가져오기
2. tr을 가져오고 그 안에서 td만 가져오기
3. 공백을 제거하기 위해 tr안의 td갯수가 적으면 제거하기

```python
url ='https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page='

for page in range(1,2):
    res = requests.get(url + str(page))
    soup = BeautifulSoup(res.content, 'html.parser')
    tr = soup.find('table', class_='type_2').find('tbody').find_all('tr')
    for td in tr:
        item = td.find_all('td')
        
        if len(item) <= 1:  # 의미 없는 데이터는 skip
            continue
       
        data = [x.get_text().strip() for x in item]
        print(data)
        
    
    
    
```

### csv 파일로 저장하기

```python
url ='https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page='

filename = '시가총액 1-50.csv'
f = open(filename, 'w', encoding='utf-8-sig', newline="")  # newline='' 은 줄과 줄 사이 공백을 제거해줌 #한글깨짐 해결 sig
writer= csv.writer(f)

title = 'N,종목명,현재가,전일비,등락률,액면가,시가총액,상장주식수,외국인비율,거래량,PER,ROE'.split(',')
writer.writerow(title)

for page in range(1,2):
    res = requests.get(url + str(page))
    soup = BeautifulSoup(res.content, 'html.parser')
    tr = soup.find('table', class_='type_2').find('tbody').find_all('tr')
    for td in tr:
        item = td.find_all('td')
        
        if len(item) <= 1:  # 의미 없는 데이터는 skip
            continue
       
        data = [x.get_text().strip() for x in item]
        #print(data)
        writer.writerow(data)
    
    
```
